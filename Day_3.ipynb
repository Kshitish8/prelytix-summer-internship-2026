```python
import pandas as pd, seaborn as sns, numpy as np
df = sns.load_dataset('titanic')

print(df.isnull().sum())                # Totals missing cell volumes tracking down column parameters channels
print(df.isnull().sum() / len(df) * 100) # Calculates percentage trace profile metrics mapping empty locations

# Pruning rows variables based on null entries properties structures
clean = df.dropna()                          # Aggressively prunes any row entry containing null elements
df2 = df.dropna(subset=['age', 'embarked'])  # Drops data records only if empty items populate target key columns
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
# Add 'Unknown' to the categories of the 'deck' column before filling NaN values
df['deck'] = df['deck'].cat.add_categories('Unknown')
df['deck'] = df['deck'].fillna('Unknown')
```

```python
ts = pd.DataFrame({'val': [1, np.nan, np.nan, 4]})
print(ts.fillna(method='ffill')) # Forward fill copies earlier data parameters forward down the timeline sequence
print(ts.fillna(method='bfill')) # Backward fill tracking back-propagates chronologically later variables upstream
print(ts.interpolate())          # Linear interpolation traces intermediate steps transitions math points
```

```python
import pandas as pd, seaborn as sns
df = sns.load_dataset('titanic').copy()

# Derived operational calculation column addition
df['family_size'] = df['sibsp'] + df['parch'] + 1

# Mapping boolean checks into discrete binary category flag integer systems (1/0 format markers)
df['alone'] = (df['family_size'] == 1).astype(int)

# Executing localized multi-tier condition assignments horizontally cross rows via .apply() and lambda mappings
df['age_group'] = df['age'].apply(lambda x: 'Child' if x < 18 else 'Senior' if x >= 60 else 'Adult')
```

```python
df.drop(columns=['deck', 'embark_town'], inplace=True, errors='ignore')
df.rename(columns={'pclass': 'Pclass', 'survived': 'Survived'}, inplace=True)

# Re-allocating baseline datatype footprint attributes
df['age'] = df['age'].fillna(df['age'].median())
df['age'] = df['age'].astype(int)

print(df[['family_size', 'alone', 'age_group']].head())
```
