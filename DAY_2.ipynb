```python
import numpy as np
data = np.array([15, 42, -3, 87, 12, 55, 0, 33])

pos_small = data[(data > 0) & (data < 50)]  # Vectorized logic bitwise AND
print(pos_small)                            # Output: [15, 42, 33]

neg_or_big = data[(data < 0) | (data > 80)] # Vectorized logic bitwise OR
print(neg_or_big)                           # Output: [-3, 87]

not_zero = data[~(data == 0)]               # Vectorized logic bitwise NOT inversion flag
print(not_zero)
```

```python
print(np.any(data < 0))            # True (Evaluates if single element matches threshold logic)
print(np.all(data > 0))            # False (Checks if complete population complies with logic checkpoint)
# print(np.all(data > 0, axis=0))  # Axis evaluations trace capability variant across 2D arrays
```

```python
idx = np.nonzero(data > 40)
print(idx)       # Generates coordinates tuple tracking matching item positions
print(data[idx]) # Output: [42, 87, 55]
```

```python
import numpy as np
scores = np.array([45, 82, 67, 55, 91, 38, 74])

# Basic if-else logic element replacement
clipped = np.where(scores >= 60, scores, 60)

# Nested np.where statement arrays tracking string replacement categories
labels = np.where(scores >= 80, 'Dist', np.where(scores >= 60, 'Merit', 'Pass'))
print(labels)

# Inline math scaling executions triggered via specific array mask evaluations
bonus = np.where(scores >= 80, scores * 1.1, scores)
print(bonus.astype(int))

# Multi-condition map list assignment using np.select
conds = [scores >= 80, scores >= 60, scores >= 40]
vals = ['A', 'B', 'C']
grades = np.select(conds, vals, default='D')
print(grades)
```

```python
import numpy as np
A = np.array([[2., 1.], [1., 3.]])

# Matrix Inversion computation
Ai = np.linalg.inv(A)
print(A @ Ai) # Output maps out a standard identity matrix footprint block

# Determinant evaluations
print(np.linalg.det(A))

# Eigenvalues & Eigenvectors Extraction
vals, vecs = np.linalg.eig(A)
print(vals)
print(vecs)

# Linear Equations Systems Solver engine tracking (Solving Ax = b structural variables equations)
b = np.array([8., 13.])
x = np.linalg.solve(A, b)
print(x)
print(A @ x) # Check matching response maps directly back to check value vector variables b

# Array Norm extraction types
v = np.array([3., 4.])
print(np.linalg.norm(v))        # Euclidean L2 norm validation calculation (5.0 value summary)
print(np.linalg.norm(A, 'fro')) # Matrix structural trace Frobenius Norm properties summary
```
