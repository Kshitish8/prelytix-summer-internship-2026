```python
import pandas as pd
import seaborn as sns

# Load dataset
df = sns.load_dataset('titanic')

# Task 1: Sort by fare descending and get top 10
top_10_paying = df.sort_values(by='fare', ascending=False).head(10)
print(top_10_paying[['class', 'fare', 'who']])
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

# Task 2: Sort by pclass ascending, then age descending
sorted_df = df.sort_values(by=['pclass', 'age'], ascending=[True, False])

# Show youngest in each class by grouping by pclass and taking the minimum age index
youngest_each_class = df.loc[df.groupby('pclass')['age'].idxmin()]
print(youngest_each_class[['pclass', 'age', 'sex', 'who']])
```

```python
import pandas as pd

# Task 3: Create a 10-student marks DataFrame with explicit ties
student_data = {
    'Student': [f'Stu_{i}' for i in range(1, 11)],
    'Marks': [85, 92, 78, 85, 92, 64, 85, 90, 78, 99]
}
students_df = pd.DataFrame(student_data)

# Apply all 5 rank methods
methods = ['average', 'min', 'max', 'dense', 'first']
for m in methods:
    students_df[f'rank_{m}'] = students_df['Marks'].rank(ascending=False, method=m)

print(students_df.sort_values(by='rank_dense'))
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

# Task 1: Compute survival rates
rate_sex = df.groupby('sex')['survived'].mean()
rate_class = df.groupby('pclass')['survived'].mean()
rate_combo = df.groupby(['sex', 'pclass'])['survived'].mean().unstack()

print("Survival by Sex:\n", rate_sex)
print("\nSurvival by Class:\n", rate_class)
print("\nSurvival by Sex x Class Combination:\n", rate_combo)
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

# Task 5: Group by embarked port with multi-column aggregate mappings
port_analysis = df.groupby('embarked').agg(
    total_passengers=('survived', 'count'),
    survival_rate=('survived', 'mean'),
    average_fare=('fare', 'mean')
).sort_values(by='survival_rate', ascending=False)

print(port_analysis)
```

```python
import pandas as pd

# Task 1: Create Relational Tables
students_table = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Carol', 'Dave']
})

grades_table = pd.DataFrame({
    'id': [1, 2, 3, 5],
    'subject': ['Math', 'Sci', 'Eng', 'Hist'],
    'marks': [88, 75, 92, 65]
})

# Execute all 4 join configurations
for join_type in ['inner', 'left', 'right', 'outer']:
    merged_res = pd.merge(students_table, grades_table, on='id', how=join_type)
    print(f"Join Type: {join_type.upper()} | Row Count: {len(merged_res)}")
    print(merged_res, "\n" + "-"*40)
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

# Task 1: Multi-metric pivot grid tracking explicit statistical criteria
pivot_grid = pd.pivot_table(
    df,
    values='age',
    index='pclass',
    columns='survived',
    aggfunc=['mean', 'std'],
    margins=True
)
print(pivot_grid.round(2))
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

# Task 2: Percentage survival rate mapped cross relational parameters
cross_tab_pct = pd.crosstab(
    index=[df['sex'], df['embarked']],
    columns=df['survived'],
    normalize='index'
) * 100

print(cross_tab_pct.round(1))
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic').copy()

# Task 2: Row-wise custom classifier lambda function mapping axis=1
def compute_outcome(row):
    status = 'Survived' if row['survived'] == 1 else 'Died'
    bracket = 'Rich' if row['fare'] > 50 else 'Poor'
    return f"{status}-{bracket}"

df['Outcome'] = df.apply(compute_outcome, axis=1)
print(df['Outcome'].value_counts())
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic').copy()

# Task 3: Compare standard equi-width bands vs quantile-based bands
df['equal_width_age'] = pd.cut(df['age'], bins=5)
df['quantile_age'] = pd.qcut(df['age'], q=5)

print("--- Equal Width Age Distribution ---")
print(df['equal_width_age'].value_counts())
print("\n--- Quantile Based Age Distribution ---")
print(df['quantile_age'].value_counts())
```

```python
import pandas as pd

# Task 1: Generate a simulated raw dataframe with dirty text layout parameters
messy_names = pd.DataFrame({
    'Raw_Name': ['   alice SMITH ', ' bOB   jones', 'carol   LEE   ']
})

# Standardize data traces across array series
messy_names['Clean_Name'] = messy_names['Raw_Name'].str.strip().str.replace(r'\s+', ' ', regex=True).str.title()

# Extract segments cleanly out to separate tracking series elements
messy_names['First_Name'] = messy_names['Clean_Name'].str.split(' ').str[0]
messy_names['Last_Name'] = messy_names['Clean_Name'].str.split(' ').str[1]

print(messy_names)
```

```python
import pandas as pd
import seaborn as sns

# Task 1: Complete pipeline parsing transformation steps to storage files
df = sns.load_dataset('titanic')

# Clean missing parameters entries fields
df['age'] = df['age'].fillna(df['age'].median())

# Multi-tab export using ExcelWriter engine structures
with pd.ExcelWriter('titanic_compiled_report.xlsx') as writer:
    df.to_excel(writer, sheet_name='Cleaned_Data', index=False)
    df.describe().to_excel(writer, sheet_name='Statistical_Summary')

print("Successfully compiled and saved 'titanic_compiled_report.xlsx'")
```
